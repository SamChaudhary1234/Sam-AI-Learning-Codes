Building a chat experience with `gr.ChatInterface`. Covers the chat callback pattern, streaming responses, system-message context for a clothing-store assistant, and dynamic system-message injection based on user input.

## Setup

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
```

```python
openai = OpenAI()
MODEL = 'gpt-4.1-mini'
system_message = "You are a helpful assistant"
```

## The chat callback

`gr.ChatInterface` calls a function with `(message, history)` on each turn — `history` is a list of `{'role', 'content'}` dicts.

```python
def chat(message, history):
    return "Hello"
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

```python
def chat(message, history):
    return f"You said '{message}', and the history has {len(history)} prior turns."
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

## A real chat callback wired to GPT

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = (
        [{"role": "system", "content": system_message}]
        + history
        + [{"role": "user", "content": message}]
    )
    response = openai.chat.completions.create(model=MODEL, messages=messages)
    return response.choices[0].message.content
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

## Streaming version of the same chat

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = (
        [{"role": "system", "content": system_message}]
        + history
        + [{"role": "user", "content": message}]
    )
    stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

## Using a richer system prompt — clothing-store assistant

Adds context about a sales event. This is one-shot prompting — the example response shapes the assistant's tone.

```python
system_message = (
    "You are a helpful assistant in a clothes store. You should try to gently encourage "
    "the customer to try items that are on sale. Hats are 60% off, and most other items are 50% off. "
    "For example, if the customer says 'I'm looking to buy a hat', "
    "you could reply something like, 'Wonderful — we have lots of hats — including several that are part of our sales event.' "
    "Encourage the customer to buy hats if they are unsure what to get."
)
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

```python
system_message += (
    "\nIf the customer asks for shoes, you should respond that shoes are not on sale today, "
    "but remind the customer to look at hats!"
)
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

## Dynamic system-message injection

Modify the system prompt based on the incoming message — useful for product-specific guidance.

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    relevant_system_message = system_message
    if 'belt' in message.lower():
        relevant_system_message += (
            " The store does not sell belts; if you are asked for belts, "
            "be sure to point out other items on sale."
        )

    messages = (
        [{"role": "system", "content": relevant_system_message}]
        + history
        + [{"role": "user", "content": message}]
    )
    stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
