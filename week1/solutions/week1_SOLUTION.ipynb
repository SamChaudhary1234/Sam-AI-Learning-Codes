# Week 1 — End of Week Exercise Solution

A simple technical-question explainer that exercises both the OpenAI API and a local Ollama model. Send any technical question and get a detailed walkthrough back from each model — useful as a learning aid throughout the rest of the course.

After Week 2 this becomes the foundation for adding a UI on top — for now it's a notebook-only tool.

````python
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
````

````python
MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
load_dotenv()
openai = OpenAI()
````

## The question to ask

Edit this cell to ask anything you'd like explained — Python idioms, a piece of code you don't understand, a concept from the course, etc.

````python
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
````

## Prompts and message list

````python
system_prompt = (
    "You are a helpful technical tutor who answers questions about "
    "python code, software engineering, data science and LLMs"
)
user_prompt = "Please give a detailed explanation to the following question: " + question

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

## Answer 1 — GPT-4o-mini with streaming

Streams the response token-by-token and updates the display in place.

````python
stream = openai.chat.completions.create(model=MODEL_GPT, messages=messages, stream=True)

response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```", "").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````

## Answer 2 — Llama 3.2 (local via Ollama)

````python
response = ollama.chat(model=MODEL_LLAMA, messages=messages)
reply = response['message']['content']
display(Markdown(reply))
````
