# Week 3 — Day 2: HuggingFace Pipelines

Hands-on with the HuggingFace high-level API: `pipeline()`. One function call to load a pre-trained model and run inference for a wide range of tasks.

Runs nicely on a free or low-cost T4 GPU runtime in Colab.

## What the Colab notebook covers

- **HuggingFace token setup** — create a token at https://huggingface.co/settings/tokens and add it as a Colab Secret named `HF_TOKEN`
- **`pipeline()` for common tasks:**
  - sentiment analysis
  - named entity recognition (NER)
  - question answering
  - summarization
  - translation
  - zero-shot classification
  - text generation
  - image generation (Stable Diffusion)
  - text-to-speech (Bark)
- **Device selection** — `device='cuda'` to use the GPU
- **Picking the right model** — browse https://huggingface.co/models filtered by task to choose pre-trained checkpoints
