- **Tokenizer comparisons across model families:**

  - Phi-3
  - Qwen2
  - StarCoder2 (code-specialized — note how it handles whitespace and code tokens differently)

- **Why same prompt → different token counts** for different tokenizers, and why this matters for cost / context-length budgeting
