## What the Colab notebook covers

- **`AutoModelForCausalLM.from_pretrained(...)`** — load a base or instruct model from HuggingFace
- **Quantization with `bitsandbytes`** — `BitsAndBytesConfig` for 4-bit `nf4` loading so a 7-8B model fits on a T4 (16GB VRAM)
- **`device_map="auto"`** — let `accelerate` distribute layers across available devices
- **Manual generation:**
  - tokenize → `model.generate(...)` → decode
  - `TextStreamer` for token-by-token streaming output
  - `max_new_tokens`, `do_sample`, `temperature`, `top_p`, `top_k` parameters
- **Models exercised:** Llama 3.1 8B Instruct, Phi-3-mini, Gemma 2 2B, Qwen2 7B, Mixtral 8x7B (with quantization)
- **Inspecting model architecture** — print `model` to see layer structure, useful as a primer for QLoRA in week 7
