# Week 5 — Day 3: LangChain RAG Pipeline

End-to-end RAG using LangChain 1.0:
- Connect to the Chroma vector store built in day 2
- Retrieve relevant chunks for each user question
- Inject them into the system prompt
- Stream the answer back through a Gradio chat UI

## Setup

```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_chroma import Chroma
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_huggingface import HuggingFaceEmbeddings
import gradio as gr
```

```python
MODEL = "gpt-4.1-nano"
DB_NAME = "vector_db"
load_dotenv(override=True)
```

## Connect to the Chroma vector store

Use the same embedding model the store was built with — `all-MiniLM-L6-v2`.

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = Chroma(persist_directory=DB_NAME, embedding_function=embeddings)
```

## Retriever and LLM

**Temperature:** controls token sampling. `temperature=0` always picks the highest-probability token (deterministic-ish), higher values sample more broadly. Note: `temperature=0` doesn't guarantee reproducibility without also fixing the random seed.

```python
retriever = vectorstore.as_retriever()
llm = ChatOpenAI(temperature=0, model_name=MODEL)
```

## Quick test of each piece

```python
retriever.invoke("Who is Avery?")
```

```python
llm.invoke("Who is Avery?")
```

## Wire them together — the RAG function

```python
SYSTEM_PROMPT_TEMPLATE = """
You are a knowledgeable, friendly assistant representing the company Insurellm.
You are chatting with a user about Insurellm.
If relevant, use the given context to answer any question.
If you don't know the answer, say so.
Context:
{context}
"""
```

```python
def answer_question(question: str, history):
    docs = retriever.invoke(question)
    context = "\n\n".join(doc.page_content for doc in docs)
    system_prompt = SYSTEM_PROMPT_TEMPLATE.format(context=context)
    response = llm.invoke([
        SystemMessage(content=system_prompt),
        HumanMessage(content=question)
    ])
    return response.content
```

```python
answer_question("Who is Averi Lancaster?", [])
```

## Gradio chat UI

```python
gr.ChatInterface(answer_question).launch()
```
