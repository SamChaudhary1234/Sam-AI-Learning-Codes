# Week 5 — Day 4: Evaluating the RAG Pipeline

Building accurate RAG isn't just about architecture and frameworks — **evaluations** are what actually tell you whether the system works for your business.

This notebook explores the test set and demonstrates two layers of evaluation:
1. **Retrieval evaluation** — how good is the vector search at surfacing the right chunks? (MRR, nDCG, keyword coverage)
2. **Answer evaluation** — given the retrieved chunks, how good is the final answer? (LLM-as-judge scoring on accuracy, completeness, relevance)

## Retrieval evaluation

Metrics returned for each question:
- **MRR** (Mean Reciprocal Rank) — how high in the ranking does the first relevant chunk appear?
- **nDCG** (Normalized Discounted Cumulative Gain) — graded relevance ranking quality
- **Keyword coverage** — what fraction of expected keywords appear in the retrieved chunks?

```python
from evaluation.eval import evaluate_retrieval, evaluate_answer
```

```python
evaluate_retrieval(example)
```

## Answer evaluation (LLM-as-judge)

Run the full RAG pipeline on the question, then have an LLM grade the answer against the reference on accuracy, completeness, and relevance (1-5 scale).

```python
eval, answer, chunks = evaluate_answer(example)
```
