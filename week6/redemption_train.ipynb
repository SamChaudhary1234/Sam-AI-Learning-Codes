# Week 6 — Optional Extra: Deep Neural Network (Training)

Train the deep residual network defined in `pricer/deep_neural_network.py` from scratch on the full 800K-item dataset.

**Heads up:** ~4 hours on an M1 Mac GPU (MPS). 5 epochs, cosine annealing learning rate.

Use the companion `redemption_run.ipynb` to load the saved weights and run inference.

```python
import os
from dotenv import load_dotenv
from huggingface_hub import login
from pricer.evaluator import evaluate
from pricer.deep_neural_network import DeepNeuralNetworkRunner
from pricer.items import Item
```

```python
LITE_MODE = False
load_dotenv(override=True)
hf_token = os.environ['HF_TOKEN']
login(hf_token, add_to_git_credential=True)
```

```python
username = "YOUR_HF_USERNAME"  # replace with your Hugging Face username
dataset = f"{username}/items_lite" if LITE_MODE else f"{username}/items_full"

train, val, test = Item.from_hub(dataset)
print(f"Loaded {len(train):,} training items, {len(val):,} validation items, {len(test):,} test items")
```

```python
runner = DeepNeuralNetworkRunner(train, val[:1000])
runner.setup()
```

```python
runner.train(epochs=5)
```

```python
def deep_neural_network(item):
    return runner.inference(item)

evaluate(deep_neural_network, test)
```

```python
runner.save('deep_neural_network.pth')
```
