# Week 6 — Results

Final comparison of every approach evaluated on the price-prediction test set this week.

Sorted by category: classical baselines (gray), human baseline (black), neural networks (orange), frontier LLMs (slateblue/green), GPT-4.1 Nano fine-tune (red).

**Lower error = better.** Among everything tried with no week-7+ work, GPT-5.1 zero-shot wins at $44.74 — beating the deep neural network ($46.49) and every frontier zero-shot baseline.

```python
import plotly.graph_objects as go

results = [
    ("Constant", "gray", 106.18),
    ("Linear Regression", "gray", 101.56),
    ("NLP + LR", "gray", 76.81),
    ("Random Forest", "gray", 72.28),
    ("XGBoost", "gray", 68.23),
    ("Human", "black", 87.62),
    ("Neural Network", "orange", 63.97),
    ("GPT 4.1 Nano", "slateblue", 62.51),
    ("Grok 4.1 Fast", "slateblue", 57.62),
    ("Gemini 3 Pro", "slateblue", 50.54),
    ("Claude 4.5 Sonnet", "slateblue", 47.10),
    ("GPT 5.1", "green", 44.74),
    ("GPT 4.1 Nano (Fine-tuned)", "red", 75.91),
    ("Deep Neural Network", "orange", 46.49),
]

labels, colors, values = zip(*results)

fig = go.Figure(go.Bar(x=labels, y=values, marker_color=colors))
fig.update_layout(
    title="Week 6 — Prediction error from each model",
    yaxis=dict(range=[0, max(values)], title="Error"),
    xaxis=dict(tickangle=-45),
    width=1000,
    height=800,
)
fig.show()
```
