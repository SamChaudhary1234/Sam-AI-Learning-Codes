# Week 7 — Day 5: Evaluation

Evaluate the fine-tuned Llama 3.2 model against held-out test data and compare against baselines (frontier LLMs, simple ML models, human baseline).

Like days 3 and 4, evaluation runs on Colab to keep the fine-tuned adapters loaded on a GPU.
