Fine-tune Llama 3.2 3B with QLoRA on the price-prediction dataset prepared in day 2.

## Why Colab

Training requires a GPU (typically a T4 or A100). The training notebook lives on Colab where the GPU runtime is available — this local notebook is just a placeholder to keep the day-by-day folder structure tidy.

## What the Colab notebook covers

- Load the prompt-formatted dataset from the Hugging Face Hub
- Load the base Llama 3.2 3B model in 4-bit quantization (`bitsandbytes`)
- Configure LoRA adapters with `peft` (target attention projections, rank, alpha, dropout)
- Set up the `SFTTrainer` from `trl` with appropriate hyperparameters (learning rate, batch size, gradient accumulation, warmup, optimizer)
- Train for a few epochs, logging loss to Weights & Biases
- Save and push the trained LoRA adapters to your Hugging Face Hub repo
