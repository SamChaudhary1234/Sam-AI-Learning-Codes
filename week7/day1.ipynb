**What QLoRA is, in one minute:**
- **LoRA** (Low-Rank Adaptation) — instead of fine-tuning all the weights of a large model, freeze the base model and train small low-rank update matrices alongside specific layers (typically attention projections). Reduces trainable parameters by 100×–1000×.
- **QLoRA** — same idea, but the frozen base model is loaded in 4-bit quantization. This lets you fine-tune large models on a single consumer GPU.
- Trainable adapters end up small (tens of MB instead of GBs), which makes them cheap to store, share, and swap.
